```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# declare path and file
data_dir = ""
excel_name = "MNRetailDatasetCSV_MNRetailDatasetCSV.xlsx"
excel_path = os.path.join(data_dir, excel_name)
```

```python
#read data into dataframe
data_df = pd.read_excel(excel_path)
```

```python
# define color
my_colors = ['red', 'yellow', 'green', 'blue', 'black', 'pink', 'orange', 'lime', 'navy', 'pink']

value = data_df.loc[1:11,:][["PRODUCT", "WEIGHTED_AVERAGE_PRICE"]].sort_values("WEIGHTED_AVERAGE_PRICE", ascending=False).head(10)

ax = value.plot.bar(x='PRODUCT', y='WEIGHTED_AVERAGE_PRICE', figsize=(20, 10),color=my_colors, title="TOP 10 total product")
plt.show()
```

```python
data_df["total_price"] = data_df["STORES_WITH_ADS"] * data_df["WEIGHTED_AVERAGE_PRICE"]
```

```python
print(data_df.head(10))
```

```python

value = data_df[["Index","REPORT_DATE", "total_price"]]
value = value.groupby('REPORT_DATE', as_index=False).sum()
print(value.head(10))
```

```python
ax = value.plot.line(x='REPORT_DATE', y='total_price', figsize=(20, 10),color=my_colors, title="Profit by year")
plt.show()
```

```python
value["Moving_AVG"] = value["total_price"].rolling(window=6).mean()
print(value.head(10))
```

```python
value.plot.line(x='REPORT_DATE', figsize=(20, 10),color=my_colors, title="Profit by year")
```

```python
df = value
print(df)
x = np.arange(df['REPORT_DATE'].size)
fit = np.polyfit(x, df['total_price'], deg=1)
print ("Slope : " + str(fit[0]))
print ("Intercept : " + str(fit[1]))
#Fit function : y = mx + c [linear regression ]
fit_function = np.poly1d(fit)

#Linear regression plot
plt.plot(df['REPORT_DATE'], fit_function(x))
#Time series data plot
plt.plot(df['REPORT_DATE'], df['total_price'])

plt.xlabel('Date')
plt.ylabel('Total Price')
plt.show()
```

```python
num_step = 5
i = num_step
leaner = list(df['total_price'][0:i])
x_data = np.array(df['index']).reshape(-1,1)
y_data = np.array(df['total_price']).reshape(-1,1)
while(i < df.shape[0]):
    x_array = x_data[i-num_step:i]
    y_array = y_data[i-num_step:i]
    regr.fit(x_array,y_array)
    x_predict = x_data[i].reshape(-1,1)
    y_predict = regr.predict(x_predict)
    leaner.append(y_predict[0][0])
    i+=1
print(leaner)
df['leaner'] = leaner
```
